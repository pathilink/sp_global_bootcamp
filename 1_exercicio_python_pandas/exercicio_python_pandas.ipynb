8. Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)


```python
def calcular_pam(pressao):
    sistolica, diastolica = map(int, pressao.split('/'))
    return (sistolica + 2 * diastolica) / 3

df['PAM'] = df['Pressão sanguínea'].apply(calcular_pam)

resultado = df.groupby('Gênero')['PAM'].mean()
resultado
```

10. Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
acima_70 = df[df['Frequência cardíaca'] > 70]['Passos diários'].mean()
abaixo_70 = df[df['Frequência cardíaca'] <= 70]['Passos diários'].mean()

print('Acima de 70:', acima_70)
print('Igual ou menor de 70:', abaixo_70)
# Não, acima de 70 dão menos passos
```
