## Desafio 6

Dicionário:
* rating: nota geral do departamento
* complaints: fator categórico sobre reclamações
* privileges: acesso a benefícios
* learning: fator categórico sobre aprendizado
* raises: fator categórico sobre aumentos salariais
* critical: fator categórico sobre feedback crítico
* advance: oportunidades de crescimento

```python
df_attitude = data('attitude')
df_attitude.head()
```

```python
# média de aprendizado (learning)) por nível de motivação (rating)
df_attitude.groupby('rating')['learning'].mean()
```

```python
# combinação de rating e complaints com maior valor médio de learning
df_mean_learning = df_attitude.groupby(['rating', 'complaints'])['learning'].mean().reset_index()
df_mean_learning.sort_values('learning', ascending=False).head(1)
```

```python
# linha com maior learning
idx_max_learning = df_mean_learning['learning'].idxmax()
idx_max_learning
```

```python
# obtém a linha com o maior valor de 'learning'
df_mean_learning.loc[idx_max_learning]
```

```python
# nível médio de learning para funcionários que receberam raises acima da média
mean_raises = df_attitude['raises'].mean()
mean_raises
```

```python
df_attitude.query('raises > 64.63')['learning'].mean()
```

## Desafio 7

```python
df_titanic = data('titanic')
df_titanic.sample(5)
```

```python
# taxa de sobrevivência geral
df_titanic[['survived']].query('survived == "yes"').count() / df_titanic[['survived']].count() * 100
```

```python
# taxa de sobrevivência por classe
survived_class = df_titanic.query('survived == "yes"').groupby('class')['survived'].count().reset_index()
survived_class
```

```python
total_class = df_titanic.groupby('class').size().reset_index(name='total')
total_class
```

```python
merged = pd.merge(survived_class, total_class, on='class')
merged
```

```python
merged['survived_ratio'] = merged['survived'] / merged['total'] * 100
merged
```

```python
# sexo com maior taxa de sobrevivência
sex_survived = df_titanic.query('survived == "yes"').groupby(['sex'])['survived'].count().reset_index()
sex_survived
```

```python
sex_total = df_titanic.groupby('sex').size().reset_index(name='total')
sex_total
```

```python
merged_sex = pd.merge(sex_survived, sex_total, on='sex')
merged_sex
```

```python
merged_sex['sex_ratio'] = merged_sex['survived'] / merged_sex['total'] * 100
merged_sex
```

```python
# quantidade de adultos e crianças sobreviventes e não sobreviventes
df_titanic.groupby('age')['survived'].value_counts()
```
